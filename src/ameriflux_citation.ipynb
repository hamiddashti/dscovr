```python
import pandas as pd
import glob
import re
import warnings

warnings.filterwarnings("ignore")
```

```python
df_citations = pd.read_csv("../data/citations_for_site_data_Ameriflux.csv", skiprows=5)
site_annual_max = pd.read_csv("../outputs/site_annual_max_lue.csv")
ameriflux_info = pd.read_excel("../data/EC/Ameriflux/sites_info.xlsx")
fluxnet_info = pd.read_excel("../data/EC/fluxnet/FLX_AA-Flx_BIF_YY_20200501.xlsx")
icos_info = pd.read_csv("../data/EC/ICOS/sites_info.csv")
icos_stations = pd.read_csv("../data/EC/ICOS/stations.csv")
```

```python
names = site_annual_max["name"]
site_information = pd.DataFrame(
    columns=["name", "type", "start_date", "end_date", "doi", "network"])
for i in range(len(names)):
    name = names.iloc[i]
    ameriflux_site = ameriflux_info[ameriflux_info["Site ID"] == name]
    icos_site = icos_info[icos_info["ID"] == name]
    fluxnet_site = fluxnet_info[fluxnet_info["SITE_ID"] == name]
    if ameriflux_site.shape[0] > 0:
        type = site_annual_max["type"].iloc[i]
        years_string = ameriflux_site[
            "Years of AmeriFlux FLUXNET Data"].values[0]

        # Split the string by commas and strip any whitespace
        years_list = [int(year.strip()) for year in years_string.split(",")]
        start_date = years_list[0]
        end_date = years_list[-1]
        if name == "US-BZF":
            print(
                f"AmeriFlux - Name: {name}, Start Year: {start_date}, End Year: {end_date}"
            )

        latitude = ameriflux_site["Latitude (degrees)"].values[0]
        longitude = ameriflux_site["Longitude (degrees)"].values[0]
        doi = ameriflux_site["AmeriFlux FLUXNET DOI"].values[0]
        network = "AMERIFLUX"
    elif fluxnet_site.shape[0] > 0:

        type = site_annual_max["type"].iloc[i]
        start_date = fluxnet_site[fluxnet_site["VARIABLE"] ==
                                  "HEIGHTC_DATE"]["DATAVALUE"].values[0][0:4]
        end_date = fluxnet_site[fluxnet_site["VARIABLE"] ==
                                "HEIGHTC_DATE"]["DATAVALUE"].values[-1][0:4]
        latitude = fluxnet_site[fluxnet_site["VARIABLE"] ==
                                "LOCATION_LAT"]["DATAVALUE"].values[0]
        longitude = fluxnet_site[fluxnet_site["VARIABLE"] ==
                                 "LOCATION_LONG"]["DATAVALUE"].values[0]
        doi = fluxnet_site[fluxnet_site["VARIABLE"] ==
                           "DOI"]["DATAVALUE"].values[0]
        network = "FLUXNET"
    elif icos_site.shape[0] > 0:
        type = site_annual_max["type"].iloc[i]
        pattern = re.compile(r"(\d{4})-(\d{4})")
        matching_files = [file for file in icos_files if re.search(name, file)]
        match = pattern.search(matching_files[0])
        start_date = match.group(1)[0:4]
        end_date = match.group(2)[0:4]
        latitude = icos_info[icos_info["ID"] == name]["lat"].values[0]
        longitude = icos_info[icos_info["ID"] == name]["lon"].values[0]
        matching_ids = icos_stations[icos_stations["Id"].str.contains(name)]
        doi = matching_ids["Id"].values[0]
        network = "ICOS"
    site_information = site_information.append(
        {
            "name": name,
            "type": type,
            "start_date": start_date,
            "end_date": end_date,
            "doi": doi,
            "latitude": latitude,
            "longitude": longitude,
            "network": network,
        },
        ignore_index=True,
    )
```

```python
site_information.to_csv("../outputs/site_information.csv", index=False)
```
