```python
names = site_annual_max["name"]

citations = pd.DataFrame(columns=["name", "type", "start_date", "end_date", "doi"])

# i = 272
# name = names.iloc[i]
# type = site_annual_max["type"].iloc[272]
# site_info = sites_info[sites_info["Site ID"] == name]
# doi = site_info["AmeriFlux FLUXNET DOI"].values[0]
# start_date = site_info["Site Start"].values[0]
# end_data = site_info["Site End"].values[0]
# # df_citations[df_citations["site_id"] == name]["citation"].values[0]

for i in range(len(names)):

    name = names.iloc[i]
    type = site_annual_max["type"].iloc[i]
    if ameriflux_sites_info[ameriflux_sites_info["Site ID"] == name].empty:
        continue
    site_info = ameriflux_sites_info[ameriflux_sites_info["Site ID"] == name]
    doi = site_info["AmeriFlux FLUXNET DOI"].values[0]
    start_date = site_info["Site Start"].values[0]
    end_data = site_info["Site End"].values[0]
    citations = citations.append(
        {
            "name": name,
            "type": type,
            "start_date": start_date,
            "end_date": end_data,
            "doi": doi,
        },
        ignore_index=True,
    )
```
